```python
from scipy.spatial.distance import euclidean, hamming


def weighted_distance(movie1, movie2):
    movie1 = movie1.flatten()  # Ensure it's 1D
    movie2 = movie2.flatten()  # Ensure it's 1D

    voteCount = euclidean([movie1[0]], [movie2[0]]) * 3.0
    country = euclidean([movie1[1]], [movie2[1]]) * 2.5
    language = euclidean([movie1[2]], [movie2[2]]) * 4.0

    kw = euclidean(movie1[3:6], movie2[3:6]) * 3.5
    cast = euclidean(movie1[6:8], movie2[6:8]) * 3.0
    crew = euclidean(movie1[8:10], movie2[8:10]) * 2.0

    rating = euclidean([movie1[10]], [movie2[10]]) * 5.0
    year = euclidean([movie1[11]], [movie2[11]]) * 4.5
    mtype = hamming([movie1[12]], [movie2[12]]) * 1.5

    genres = hamming(movie1[13:], movie2[13:]) * 6.0

    return float(
        voteCount
        + country
        + language
        + kw
        + cast
        + crew
        + rating
        + year
        + mtype
        + genres
    )
```

```python
from sklearn.neighbors import NearestNeighbors

feature_columns = merged_df.columns[1:]
X = merged_df[feature_columns].values

knn = NearestNeighbors(n_neighbors=6, metric=weighted_distance)
knn.fit(X)
```

```python
def recomment(index):
    distances, indices = knn.kneighbors(merged_df.iloc[index, 1:].values.reshape(1, -1))
    recommended_indices = indices[0][1:]
    print(f'Recommendations for {details_df["title"][index]} :')

    for rec in recommended_indices:
        print(details_df["title"][rec])
```
