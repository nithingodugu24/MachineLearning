```python
import requests
import zipfile
import io
import os
import re
import shutil  # For removing folders


def get_movie_words_zip(url):
    response = requests.get(url)

    if response.status_code == 200:
        extract_path = "extracted_srt"

        # Remove and recreate directory to avoid stale files
        if os.path.exists(extract_path):
            shutil.rmtree(extract_path)
        os.makedirs(extract_path)

        # Extract new ZIP file
        with zipfile.ZipFile(io.BytesIO(response.content)) as zip_file:
            zip_file.extractall(extract_path)

        words = []
        for file_name in os.listdir(extract_path):
            try:
                if file_name.endswith(".srt") or file_name.endswith(".ass"):
                    srt_path = os.path.join(extract_path, file_name)
                    print("Found SRT file:", srt_path)

                    with open(srt_path, "r", encoding="utf-8") as file:
                        for line in file:
                            # Remove timestamps and numbers
                            line = re.sub(r"\d+[:.,]?\d*[:.,]?\d*", "", line)

                            # Extract meaningful words
                            clean_words = re.findall(r"[a-zA-Z]+", line.lower())
                            words.extend(clean_words)

                    try:
                        shutil.rmtree(extract_path)
                    except:pass
                    return " ".join(words)  # Properly remove extra spaces
            except:
                pass

    else:
        print("Error: Request failed with status code", response.status_code)
        return None
```

```python
import requests
import zipfile
import io
import os
import re
import shutil
import concurrent.futures
import pandas as pd
import tempfile


def read_subtitle_file(srt_path):
    """Read the subtitle file with auto-detected encoding."""
    encodings = ["utf-8", "latin-1", "windows-1252"]  # Common subtitle encodings

    for enc in encodings:
        try:
            with open(srt_path, "r", encoding=enc) as file:
                return file.readlines()  # Successfully read file
        except UnicodeDecodeError:
            continue  # Try next encoding

    print(f"Error: Unable to decode {srt_path} with known encodings.")
    return []  # Return empty list if all fail


# Function to download and extract words from subtitles
def get_movie_words_zip(url):
    response = requests.get(url)
    # print(response)

    if response.status_code == 200:
        with tempfile.TemporaryDirectory() as extract_path:
            try:
                # Extract ZIP file
                with zipfile.ZipFile(io.BytesIO(response.content)) as zip_file:
                    zip_file.extractall(extract_path)

                words = []
                for file_name in os.listdir(extract_path):
                    if file_name.endswith(".srt") or file_name.endswith(".ass"):
                        srt_path = os.path.join(extract_path, file_name)

                        # Use encoding detection
                        lines = read_subtitle_file(srt_path)
                        for line in lines:
                            line = re.sub(
                                r"\d+[:.,]?\d*[:.,]?\d*", "", line
                            )  # Remove timestamps
                            clean_words = re.findall(
                                r"[a-zA-Z]+", line.lower()
                            )  # Extract words
                            words.extend(clean_words)

                return " ".join(words)  # Return extracted words
            except Exception as e:
                print(f"Error processing subtitles: {e}")

    else:
        print(f"Error code: {response.status_code}")


# Function to get subtitles from IMDb ID
def get_subtitles(imdbid):
    imdbid = int(imdbid.replace("tt", ""))
    url = f"https://www.opensubtitles.org/en/search/sublanguageid-eng/imdbid-{imdbid}/rss_0_91"

    try:
        resp = requests.get(url)
        if resp.status_code != 200:
            return None

        match = re.search(r"(?<=en/subtitles/)\d+", resp.text)
        if not match:
            return None

        subid = match.group()
        # print(f"Subid: {subid}")
        return get_movie_words_zip(
            f"https://dl.opensubtitles.org/en/download/sub/{subid}"
        )

    except Exception as e:
        print(f"Error fetching subtitles for {imdbid}: {e}")
        return None


# Function to process subtitles concurrently
def process_subtitles(index, imdbid):
    words = get_subtitles(imdbid)
    return index, words
```

```python
combined_df["subtitles"] = None
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=1000) as executor:
    futures = {
        executor.submit(process_subtitles, i, combined_df["id"][i]): i
        for i in range(10,100)
    }

    for future in concurrent.futures.as_completed(futures):
        index, words = future.result()
        # print(words)
        if words:
            combined_df.at[index, "subtitles"] = words
        else:
            combined_df.drop(index, inplace=True)  # Remove row if no subtitles found

combined_df.reset_index(drop=True, inplace=True)
```
