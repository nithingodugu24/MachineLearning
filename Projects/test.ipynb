```python
import pandas as pd
```

```python
merged_df = pd.read_pickle("processed_movies.pkl")
```

```python
import pickle
knn_model = None
with open("knn_model.pkl", "rb") as f:
    knn_model = pickle.load(f)
```

```python
def recomment(index):
    distances, indices = knn_model.kneighbors([merged_df.iloc[index, 2:].to_numpy()])
    recommended_indices = indices[0][1:]
    print(f'Recommendations for {merged_df["title"][index]} :')
    for rec in recommended_indices:
        print(merged_df["title"][rec])
```
