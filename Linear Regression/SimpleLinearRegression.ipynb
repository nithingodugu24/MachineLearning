```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("placements.csv")
```

```python
df.head(4)
```

```python
plt.scatter(df["cgpa"], df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
```

#Seperating X and Y columns

```python
X = df.iloc[:,0:1]
Y = df.iloc[:, 1:2]
```

```python
from sklearn.model_selection import train_test_split

# In single way
# X_train, X_test, Y_train, Y_test = train_test_split(X,Y test_size=0.2, random_state=2)

X_train, X_test = train_test_split(X, test_size=0.2, random_state=2)
Y_train, Y_test = train_test_split(Y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.intercept_
```

```python
lr.predict([[5.88]])
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df["cgpa"], df["package"])
plt.plot(X_test, lr.predict(X_test),color="red")
plt.plot(X_train, lr.predict(X_train), color="green")
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
m = lr.coef_  #Slope
b = lr.intercept_ #Y Intercept
x_value = 5.88
y_value = m * (x_value) + b
y_value
```
